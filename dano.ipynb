```python
import pandas as pd
import numpy as np
bd = pd.read_csv('dano.csv',on_bad_lines='warn', delimiter=';')
```

```python
applications = bd.groupby('application_id').size()
applications.name = 'count'
applications
```

```python
bd = bd.join(applications, on='application_id', how='left')
bd
```

```python
bd['k'] = bd['req_amt']/bd['monthly_income_amt']
bd = bd.round({'k' : 0})
bd['k'] = bd['k'].fillna(0)
bd['k'] = bd['k'].apply(lambda x: min(x, 13.0))
bd
```

```python
db_valid_type = bd.drop(bd[bd.agreement_product_code.isin(["CC", "CL", "CLR", "CLC", "Unsuccess"])].index)
db_valid_type
```

```python
db_valid_type.pivot_table(index=['k'], columns=['count'], aggfunc='count', values=['application_id'])
```

```python
bits = bd.drop(bd[bd.agreement_product_code.isin(["CC", "To CLR", "To CLC", "Unsuccess"])].index)
bits
```

```python
bits.pivot_table(index=['age'], columns=['agreement_product_code'], aggfunc='mean', values=['account_rate'])
```
